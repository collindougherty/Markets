# Options Data

```python
import yfinance as yf
import pandas as pd
```

```python
aapl = yf.Ticker("AAPL")
current_price = aapl.info['regularMarketPrice']
dates = aapl.options
options = aapl.option_chain().calls
for i in dates:
    it_options = aapl.option_chain(date = i)
    call_options = it_options.calls
    options = options.append(call_options)
#options.calls
```

```python
options
```

```python
tradable_options = options[options.loc[:,'volume']>100]
```

```python
tradable_options['midpoint'] = .5*(tradable_options['bid']+tradable_options['ask'])
tradable_options['breakeven'] = tradable_options['midpoint'] + tradable_options['strike']
tradable_options
```

```python
tradable_options['be_move_needed'] = tradable_options['breakeven'] - current_price
tradable_options['be_percent_move_needed'] = 100*tradable_options['move_needed']/current_price
tradable_options['exercise_value'] = current_price - tradable_options['strike']
tradable_options['5pct_move'] =(current_price*1.05)-tradable_options['strike']
tradable_options
```

```python
import numpy as np
n = np.linspace(0,2, 101)
price_moves = current_price*n
exercise_vals = price_moves - 105
exercise_vals
```
