```python
import requests
import json
import pandas as pd
```

```python
urls = {'https://api.glassnode.com/v1/metrics/market/price_usd_close',
        'https://api.glassnode.com/v1/metrics/distribution/exchange_net_position_change',
        'https://api.glassnode.com/v1/metrics/indicators/reserve_risk', 
        'https://api.glassnode.com/v1/metrics/indicators/puell_multiple',
       'https://api.glassnode.com/v1/metrics/indicators/rhodl_ratio',
       'https://api.glassnode.com/v1/metrics/market/mvrv_z_score',
       'https://api.glassnode.com/v1/metrics/indicators/sopr_adjusted',
       'https://api.glassnode.com/v1/metrics/indicators/net_unrealized_profit_loss',
       'https://api.glassnode.com/v1/metrics/mining/marketcap_thermocap_ratio'}
```

```python
data = []
for url in urls:
    label = url.split('/')[-1]
    
    #make API request
    res = requests.get(url,params={'a': 'BTC', 'api_key': API_KEY})
    
    #convert to pandas dataframe
    df = pd.read_json(res.text, convert_dates=['t'])
    df.set_index('t', inplace = True)
    df.rename(columns={'v':label}, inplace= True)
    data.append(df)
```

```python
# creating a dataframe from the list of results generated from the API
data_df = pd.concat(data, axis=1)
data_df
```
